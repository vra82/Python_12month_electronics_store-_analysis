# Nachweis: Produkte unter X USD vs. Produkte über Z USD

```python
# Konvertierung zu numerischen Werten
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'], errors='coerce')
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'], errors='coerce')

# Entfernung von NaN-Werten
all_data_clean = all_data.dropna(subset=['Price Each', 'Quantity Ordered'])

# Definition von Preis-Kategorien
price_threshold_low = 50  # Unter $50
price_threshold_high = 200  # Über $200

# Segmentierung nach Preis
low_price = all_data_clean[all_data_clean['Price Each'] < price_threshold_low]
high_price = all_data_clean[all_data_clean['Price Each'] > price_threshold_high]

# Berechnung der durchschnittlichen Bestellmenge
avg_qty_low = low_price['Quantity Ordered'].mean()
avg_qty_high = high_price['Quantity Ordered'].mean()

# Verhältnis-Berechnung
ratio = avg_qty_low / avg_qty_high if avg_qty_high > 0 else 0

print(f"Produkte unter ${price_threshold_low}:")
print(f"  - Anzahl Bestellungen: {len(low_price)}")
print(f"  - Durchschnittliche Menge pro Bestellung: {avg_qty_low:.2f}")
print(f"\nProdukte über ${price_threshold_high}:")
print(f"  - Anzahl Bestellungen: {len(high_price)}")
print(f"  - Durchschnittliche Menge pro Bestellung: {avg_qty_high:.2f}")
print(f"\n🎯 FINDING: Produkte unter ${price_threshold_low} werden {ratio:.1f}× häufiger bestellt als Produkte über ${price_threshold_high}")

# Visualisierung mit Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))
categories = [f'Unter ${price_threshold_low}', f'Über ${price_threshold_high}']
quantities = [avg_qty_low, avg_qty_high]
colors = ['#2ecc71', '#e74c3c']

bars = ax.bar(categories, quantities, color=colors, alpha=0.8, edgecolor='black', linewidth=2)

# Beschriftung der Balken
for i, (bar, qty) in enumerate(zip(bars, quantities)):
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2., height,
            f'{qty:.2f}x',
            ha='center', va='bottom', fontsize=12, fontweight='bold')

# Verhältnis-Annotation
ax.text(0.5, max(quantities) * 0.8, f'Verhältnis: {ratio:.1f}×',
        ha='center', fontsize=14, fontweight='bold',
        bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))

ax.set_ylabel('Durchschnittliche Bestellhäufigkeit', fontsize=12, fontweight='bold')
ax.set_title(f'Preis-Volumen-Analyse: Günstprodukte vs. Premium-Produkte\nProdukte unter ${price_threshold_low} werden {ratio:.1f}× häufiger bestellt', 
             fontsize=13, fontweight='bold', pad=20)
ax.set_ylim(0, max(quantities) * 1.15)
plt.grid(axis='y', alpha=0.3, linestyle='--')
plt.tight_layout()
plt.savefig('price_volume_analysis.png', dpi=300, bbox_inches='tight')
plt.show()
```
