# what products are most ofter sold together?

```python
# Identify rows with duplicated 'Order Date'
df = all_data[all_data['Order Date'].duplicated(keep=False)]

# Group products by 'Order ID' and concatenate them
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

# Create a new DataFrame containing 'Order ID' and 'Grouped' columns, dropping duplicates
df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

# Loop through each row in the 'Grouped' column
for row in df['Grouped']:
    # Split the row string into a list of products
    row_list = row.split(',')
    
    # Generate combinations of two products and update the counter
    count.update(combinations(row_list, 2))
    
count
```

```python
# Group the data by product and sum the quantity ordered for each product
product_sales = all_data.groupby('Product')['Quantity Ordered'].sum()


product_sales 
```

```python
import matplotlib.pyplot as plt

# Group the data by product and sum the quantity ordered for each product
product_sales = all_data.groupby('Product')['Quantity Ordered'].sum()

# Sort products by total quantity ordered in descending order
product_sales = product_sales.sort_values(ascending=False)

# Plotting
plt.figure(figsize=(12, 6))
product_sales.plot(kind='bar')
plt.title('Total Quantity Ordered for Each Product')
plt.xlabel('Product')
plt.ylabel('Total Quantity Ordered')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
prices = all_data.groupby('Product')['Price Each'].mean()
print(prices)
```

```python
import matplotlib.pyplot as plt

# Group the data by product and calculate the total quantity ordered for each product
product_quantity = all_data.groupby('Product')['Quantity Ordered'].sum()

# Group the data by product and calculate the average price for each product
product_prices = all_data.groupby('Product')['Price Each'].mean()

# Create figure and axes objects
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot total quantity ordered on the primary y-axis
color = 'tab:blue'
ax1.set_xlabel('Product')
plt.xticks(rotation=90)
ax1.set_ylabel('Total Quantity Ordered', color=color)
ax1.bar(product_quantity.index, product_quantity.values, color=color)
ax1.tick_params(axis='y', labelcolor=color)

# Create a twin axes object for average price
ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Average Price ($)', color=color)
ax2.plot(product_prices.index, product_prices.values, color=color, marker='o', linestyle='-', linewidth=2)
ax2.tick_params(axis='y', labelcolor=color)

# Set title and labels
plt.title('Total Quantity Ordered and Average Price for Each Product')
plt.xticks(rotation=90)  # Rotate x-axis labels
plt.tight_layout()  # Adjust layout to prevent clipping of labels

# Show plot
plt.show()
```
